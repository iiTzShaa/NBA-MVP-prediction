```python
import os
import requests
import shutil
from bs4 import BeautifulSoup
import time
import pandas as pd
import warnings
from selenium import webdriver
from selenium.webdriver.chrome.service import Service  



warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
years = list(range(1991,2024))
url_start = "https://www.basketball-reference.com/awards/awards_{}.html"
for year in years:
    try:
        url = url_start.format(year)
        response = requests.get(url)
        
        if response.status_code == 429:  
            retry_after = int(response.headers.get('Retry-After', 60)) 
            print(f"Rate limit exceeded. Retrying after {retry_after} seconds...")
            time.sleep(retry_after)
            response = requests.get(url)
        
        response.raise_for_status()

        os.makedirs("mvp", exist_ok=True)

        with open(f"mvp/{year}.html", "w+", encoding="utf-8") as f:
            f.write(response.text)
        print(f"Saved {year} MVP page successfully.")

        time.sleep(2)

    except requests.exceptions.RequestException as e:
        print(f"Failed to retrieve data for {year}: {e}")
```

```python
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

```python
for year in years:
    url = team_stats_url.format(year)
    response = requests.get(url)
    if response.status_code == 200:
        with open(f"teams/{year}.html", "w+", encoding="utf-8") as f:
            f.write(response.text)
        print(f"Successfully saved data for {year}")
    else:
        print(f"Failed to retrieve data for {year}: {response.status_code}")
    time.sleep(5)  # Wait for 5 seconds before the next request

```

```python
df_teams= []
years = list(range(1991,2024))
for year in years:
    with open("teams/{}.html".format(year), "r", encoding="utf-8") as f:
        page= f.read()
    soup = BeautifulSoup(page,"html.parser")        
   # Eastern Conference
    e_team_table = soup.find(id="div_divs_standings_E")
    for tr in e_team_table.find_all('tr', class_='thead'):
        tr.decompose()  # Remove the header rows
    e_team = pd.read_html(str(e_team_table))[0]
    e_team["Year"] = year
    e_team["Team"] = e_team["Eastern Conference"]
    del e_team["Eastern Conference"]
    df_teams.append(e_team)

    # Western Conference
    w_team_table = soup.find(id="div_divs_standings_W")
    for tr in w_team_table.find_all('tr', class_='thead'):
        tr.decompose()  
    w_team = pd.read_html(str(w_team_table))[0]
    w_team["Year"] = year
    w_team["Team"] = w_team["Western Conference"]
    del w_team["Western Conference"]
    df_teams.append(w_team)
```

```python
df_teams
```

```python
teams = pd.concat(df_teams)
```

```python
teams.to_csv("teams.csv")
```
